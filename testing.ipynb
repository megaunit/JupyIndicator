# 20 Newsgroups Text Classification

This notebook explores text classification on 5 categories from the 20 Newsgroups dataset:
- `alt.atheism`
- `talk.religion.misc`
- `comp.graphics`
- `sci.space`
- `rec.autos`

We'll walk through data loading, preprocessing, vectorization, and a baseline model.

## 1. Imports and Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

## 2. Load the Dataset

We fetch both `train` and `test` subsets, stripping headers/footers/quotes to force the model to learn from body text only.

```python
categories = [
    'alt.atheism',
    'talk.religion.misc',
    'comp.graphics',
    'sci.space',
    'rec.autos',
]
remove = ('headers', 'footers', 'quotes')

train = fetch_20newsgroups(subset='train', categories=categories, remove=remove)
test = fetch_20newsgroups(subset='test', categories=categories, remove=remove)

print(f'Train documents: {len(train.data)}')
print(f'Test documents:  {len(test.data)}')
print(f'Categories:      {train.target_names}')
```

### Class distribution

A quick look at how balanced the training set is across the 5 categories.

```python
counts = pd.Series(train.target).value_counts().sort_index()
counts.index = [train.target_names[i] for i in counts.index]
print(counts)

counts.plot(kind='bar', title='Training documents per category')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## 3. Peek at a Sample Document

```python
idx = 0
print(f'Label: {train.target_names[train.target[idx]]}')
print('-' * 60)
print(train.data[idx][:500])
```

## 4. TF-IDF Vectorization

We build a TF-IDF matrix with `max_features=5000`, matching the setup used in the main assignment notebook.

```python
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X_train = vectorizer.fit_transform(train.data)
X_test = vectorizer.transform(test.data)
y_train, y_test = train.target, test.target

print(f'X_train shape: {X_train.shape}')
print(f'X_test shape:  {X_test.shape}')
print(f'Vocab size:    {len(vectorizer.vocabulary_)}')
```

## 5. Baseline: Multinomial Naive Bayes

Fit a simple Naive Bayes classifier on TF-IDF features as a baseline.

```python
clf = MultinomialNB()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print(f'Test accuracy: {acc:.4f}')
```

### Per-class performance

```python
print(classification_report(y_test, y_pred, target_names=train.target_names))
```

### Confusion matrix

```python
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm, index=train.target_names, columns=train.target_names)
print(cm_df)
```
